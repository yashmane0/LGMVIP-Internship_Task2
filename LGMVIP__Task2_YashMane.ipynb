# **Image to Pencil Sketch with Python - Task 2**
The aim is to programmatically transform a color image into a back and white pencil sketch with enhanced details, we first read the image in RGB format. Convert the image to grayscale, removing color information Then create an inverted grayscale image by reversing the intensity values. Blur the inverted image to enhance artistic appearance. Blend the grayscale and inverted blurry images by dividing corresponding pixel values. The resulting image will resemble a pencil sketch.


# **Importing necessary libraries**

```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Read Image
image=cv2.imread("/content/reynaknifethumbnail.png")
```

```python
image=cv2.imread("/content/reynaknifethumbnail.png")
plt. imshow (cv2.cvtColor (image, cv2.COLOR_BGR2RGB))
plt.axis ('off')
plt.show()
```

# **Coverting the original image into grey scale**

```python
grayImage = cv2.cvtColor (image, cv2.COLOR_BGR2GRAY)
grayImage
```

```python
plt.imshow (grayImage)
```

# **Inverting the grey scale image**

```python
invertedImage = 255-grayImage
plt.imshow(cv2.cvtColor (invertedImage, cv2.COLOR_BGR2RGB))
```

# **Blur the image using the Gaussian Function in OpenCV**

```python
blurred = cv2.GaussianBlur (invertedImage, (21,21), 0)
plt.imshow (blurred)
```

# **Inverting the above blured image to get pencil sketch**

```python
invertedBlur = 255 - blurred
pencilsketch = cv2.divide(grayImage, invertedBlur, scale=256.0)
plt.imshow(pencilsketch)
```

# **Final Output: Pencil Sketch**

```python
invertedBlur=255-blurred
pencilsketch=cv2.divide (grayImage, invertedBlur, scale=256.0)
plt.imshow (cv2.cvtColor (pencilsketch, cv2.COLOR_BGR2RGB))
plt.axis ('off')
plt.show()
```
